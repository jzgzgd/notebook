```python
import pandas as pd
import re

# 读取数据
fifa = pd.read_csv('./fifa21_raw_data.csv')

# 假设你要分割的数据在名为 'Team_Contract' 的列中
# 使用正则表达式来分割字符串并创建新的列
pattern = re.compile(r'(?P<Team>[\w\s]+)(?P<Contract>\d{4} ~ \d{4})')
extracted = fifa['Team & Contract'].str.extract(pattern)

# 将新的列添加到数据表中
fifa = pd.concat([fifa, extracted], axis=1)
```

```python
fifa
```

```python
fifa['Team']=fifa['Team'].str.replace('\n','')
```

```python
fifa['Contract']=fifa['Contract'].str.replace('\n','')
```

```python
fifa.drop(columns={'Team & Contract'},inplace=True)
```

Convert Inch to CM

```python
def convert_to_cm(height_str):
    match = re.match(r"(\d+)'(\d+)\"", height_str)
    if match:
        feet, inches = map(int, match.groups())
        return (feet * 30.48) + (inches * 2.54)
    else:
        return None  # 或其他的错误处理方式


# 使用 apply 方法将转换函数应用到 "Height" 列
fifa['Height'] = fifa['Height'].apply(convert_to_cm)
```

```python
fifa.head()
```

```python
def convert_weight(weight_str):
    # 移除'lbs'，并将结果转换为整数
    return int(weight_str.replace('lbs', ''))

fifa['Weight'] = fifa['Weight'].apply(convert_weight)
```

```python
def ConvertMoney(money:str):
    if '€' in money:
        money=money.replace("€","")
    if 'M' in money:
        return int(float(money.replace('M',''))*1000000)
    elif 'K' in money:
        return int(float(money.replace('K',''))*1000)

fifa['Value'] = fifa['Value'].apply(ConvertMoney)
fifa['Wage'] = fifa['Wage'].apply(ConvertMoney)
fifa['Release Clause'] = fifa['Release Clause'].apply(ConvertMoney)
fifa[['Value','Wage','Release Clause']]
```
